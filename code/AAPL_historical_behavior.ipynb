```python
import yfinance as yf
import numpy as np
import pandas as pd

def calculate_market_params(ticker_symbol="AAPL"):
    # We Calculate Historical Volatility (sigma)
    # We pull 1 year of data to get a stable estimate
    stock = yf.Ticker(ticker_symbol)
    hist = stock.history(period="1y")
    
    # Calculate daily log returns
    hist['Log_Ret'] = np.log(hist['Close'] / hist['Close'].shift(1))
    
    # And then standard deviation of daily returns
    daily_vol = hist['Log_Ret'].std()
    
    # Then we annualize volatility (252 trading days in a year)
    sigma = daily_vol * np.sqrt(252)
    
    # Get Risk-Free Rate (r)
    # By the PINN paper, ^IRX is the ticker for the 13-week US Treasury Bill yield
    tnx = yf.Ticker("^IRX")
    r_percent = tnx.history(period="1d")['Close'].iloc[-1]
    r = r_percent / 100 # Convert from e.g., 4.2 to 0.042
    
    return r, sigma

# Execute!!!
r_val, sigma_val = calculate_market_params("AAPL")

print(f"--- Derived Parameters for {datetime.now().strftime('%Y-%m-%d')} ---")
print(f"Risk-Free Rate (r): {r_val:.4f}")
print(f"Historical Volatility (sigma): {sigma_val:.4f}")
```
